```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0, 5, 11)
y = x**2
```

```python
# 1. Functional Method
plt.plot(x,y)
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y, x) # x축 y값
```

```python
# 2. object-oriented method
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
fig
```

```python
#ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 상대적인 위치 1이하 값, 시작 x점(left bottom) 가로 세로 길이 width height
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 시작 x점 (0.2, 0.5) 길이 width height
#ax1.plot(x,y)
ax2.plot(y, x)

#ax1.set_title('Large plot')
ax2.set_title('Smaller plot')
fig
```

### 3.쪼개기 subplot + object

```python
# 3.쪼개기 subplot + object
fig, axes = plt.subplots(nrows=3, ncols=2)
plt.tight_layout() #각 챠트들이 겹치지 않도록 해주는 정말 고마운 함수다.
#출처: https://enjoyiot.tistory.com/entry/Visualization-with-Matplotlib-and-Pandas-2-Matplotlib-활용하기 [Shining...]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
[ currnt_ax.plot(x,y) for currnt_ax in axes]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4)) # figsize 단위 인치
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout() # 항상 마지막에 이용
```

```python
# legend  범례  # 한 차트에 여러개의 그래프 표현, 색상 or marker로 구분 # legend 편리
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=10) # legend 표시
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=(0.1, 0.9)) # 상대수치로도 표현가능 x 10%, y 90%지점
```

```python
# plot 모양새 바꾸기
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 
        color='purple', linewidth=4, alpha=0.5, linestyle=':', 
# purple,green,blue,red   # lw                  # ls # '-' (일직선) '--' (점선)  '-.'   ':' 삼각형
        marker='^', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
    #  '+' 수직 바  'o' 동그라미   's'  사각형   '1' 작은 점 크기순('1', '2',,,,,)
```

```python
# 축의 데이터의 표시될 구간 변경, set_xlim, set_ylim
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
       color='purple', linewidth=4, alpha=0.5, linestyle=':', 
# purple,green,blue,red   # lw                  # ls # '-' (일직선) '--' (점선)  '-.'   ':' 삼각형
        marker='^', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
    #  '+' 수직 바  'o' 동그라미   's'  사각형   '1' 작은 점 크기순('1', '2',,,,,))
ax.set_xlim([0,1]) # 0 ~ 1 사이의 데이터만 보여줌
ax.set_ylim([0,1]) # 
```
