```python
## reset_index()
# 기존 인덱스는 index라는(set_index()에 명명시 그 이름) column으로 추가
   # drop = False (default) => 기존 index 추가
   # drop = True            => 기존 index 삭제 
# inplace = False  => 새로운 df return
# inplace = True  => 원본 df 수정, return None

titanic_df = pd.read_csv('titanic_train.csv')
titanic_reset_df = titanic_df.reset_index(inplace=False)
print( titanic_reset_df.head(3) )

value_counts = titanic_df['Pclass'].value_counts() # series
new_value_counts = value_counts.reset_index(inplace=False)
new_value_counts.head(3)
```
